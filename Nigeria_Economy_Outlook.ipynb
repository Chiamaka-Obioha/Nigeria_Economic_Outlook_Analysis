```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
Africa_economic_outlook1 = Africa_economic_outlook.fillna(0)
```

```python
Africa_economic_outlook1.isna().sum()
```

```python
Nigeria_economic_outlook = Africa_economic_outlook1.loc[1104:1132, :]
```

```python
Nigeria_economic_outlook
```

```python
Nigeria_economic_outlook1 = Africa_economic_outlook1[Africa_economic_outlook1['Country and Regions Name'] == "Nigeria"]
```

```python
Nigeria_economic_outlook1
```

```python
Nigeria_economic_outlook2 = Nigeria_economic_outlook1.T # you can also use .transpose()
```

```python
Nigeria_economic_outlook2
```

```python
Nigeria_economic_outlook2.iloc[[4]]
```

```python
Nigeria_economic_outlook2.columns = Nigeria_economic_outlook2.iloc[4]
```

```python
Nigeria_economic_outlook2
```

```python
Nigeria_economic_outlook3 = Nigeria_economic_outlook2.drop(['Country and Regions', 'Country and Regions Name', 'Country and Regions - RegionId', 'Indicators', 'Indicators Name', 'Scale', 'Units'], axis=0)
```

```python
Nigeria_economic_outlook3
```

```python
Nigeria_economic_outlook4 = Nigeria_economic_outlook3.reset_index()
```

```python
Nigeria_economic_outlook4
```

```python
Nigeria_economic_outlook5 = Nigeria_economic_outlook4.rename({'index': 'Year'}, axis=1)
```

```python
Nigeria_economic_outlook5 
```

```python
Nigeria_economic_outlook5.plot(x= 'Year', y='Real GDP growth (annual %)', style='o')
plt.title('Year vs Real GDP growth (annual %)')
plt.xlabel('Year')  
plt.ylabel('Real GDP growth (annual %)')  
plt.show()
```

```python
seabornInstance.histplot(Nigeria_economic_outlook5['Real GDP growth (annual %)'])
```

```python
X = Nigeria_economic_outlook5[('Year')].values.reshape(-1,1)
y = Nigeria_economic_outlook5[('Real GDP growth (annual %)')].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
df1 = df.head(30)
df1.plot(kind='bar',figsize=(10,5))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='purple')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
